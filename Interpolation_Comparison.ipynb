```python

% Figure 2: Interpolation Comparison for Q4 2025
clc;
clear;
close all;
% Known RPPI data
x = [1 2 3];
y = [180.4258 189.3172 180.6855];
% Missing quarter
xq = 4;
% Smooth plotting range
xx = 1:0.01:4;
% Linear interpolation
y_linear_q4 = interp1(x, y, xq, 'linear', 'extrap');
y_linear_curve = interp1(x, y, xx, 'linear', 'extrap');
% Lagrange interpolation (quadratic for 3 points)
p_lagrange = polyfit(x, y, 2);
y_lagrange_q4 = polyval(p_lagrange, xq);
y_lagrange_curve = polyval(p_lagrange, xx);
% Cubic spline interpolation
y_spline_q4 = spline(x, y, xq);
y_spline_curve = spline(x, y, xx);
figure;
plot(xx, y_linear_curve, 'LineWidth', 2);
hold on;
plot(xx, y_lagrange_curve, 'LineWidth', 2);
plot(xx, y_spline_curve, 'LineWidth', 2);
% Known data points
scatter(x, y, 80, 'k', 'filled');
% Estimated Q4 points
scatter(xq, y_linear_q4, 80, 'filled');
plot(xx, y_lagrange_curve,'--','LineWidth',2) % dashed line
plot(xx, y_spline_curve,':','LineWidth',2)    % dotted line
grid on;
xticks([1 2 3 4]);
xticklabels({'Q1 2025','Q2 2025','Q3 2025','Q4 2025'});
xlabel('Quarter');
ylabel('RPPI Value');
title('Comparison of Interpolation Methods for Estimating Q4 2025 RPPI');
legend('Linear Interpolation', ...
      'Lagrange Polynomial Interpolation', ...
      'Cubic Spline Interpolation', ...
      'Known Data Points', ...
      'Estimated Q4 Points', ...
      'Location', 'best');

```
