```python
clc;
clear;
close all;
% Known 2025 values including interpolated Q4
x = [1 2 3 4];
y = [180.4258 189.3172 180.6855 172.0538];
% Forecast range up to Q4 2027
xf = 1:12;
% Quarter labels
labels = {'Q1 2025','Q2 2025','Q3 2025','Q4 2025', ...
         'Q1 2026','Q2 2026','Q3 2026','Q4 2026', ...
         'Q1 2027','Q2 2027','Q3 2027','Q4 2027'};
% Linear approximation
p1 = polyfit(x, y, 1);
y_linear = polyval(p1, xf);
% Quadratic approximation
p2 = polyfit(x, y, 2);
y_quad = polyval(p2, xf);
% Cubic spline approximation
y_spline = spline(x, y, xf);
figure;
plot(xf, y_linear, '-o', 'LineWidth', 2, 'MarkerSize', 6);
hold on;
plot(xf, y_quad, '-s', 'LineWidth', 2, 'MarkerSize', 6);
plot(xf, y_spline, '-^', 'LineWidth', 2, 'MarkerSize', 6);
% Known/interpolated points
scatter(x, y, 90, 'k', 'filled');
grid on;
xticks(1:12);
xticklabels(labels);
xtickangle(45);
xlabel('Quarter');
ylabel('RPPI Value');
title('RPPI Forecast Using Approximation Methods (2026–2027)');
legend('Linear Approximation', ...
      'Quadratic Approximation', ...
      'Cubic Spline Approximation', ...
      'Observed / Interpolated Data', ...
      'Location', 'best');

```
